# 1. Gaussian kernel 그리기
# $f_1 = exp(-\frac{||x-l^{(1)}||^2}{2\sigma^2})$
# $f_1 = exp(-\frac{||v_1-v_2||^2}{2\sigma^2})$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#vector v1과 v2의 차이에 대한 Gaussian kernel을 구하자
#val = GaussianKernel(np.array([0,0]), np.array([3,3]), 1.0)
def GaussianKernel(v1, v2, sigma):
    res=np.exp(-np.sum([np.power(v1,2),np.power(v2,2),-2*v1*v2])/(2*np.power(sigma,2)))
    return res
```

```python
val = GaussianKernel(np.array([0,0]), np.array([3,5]), 1.0)
print val
```

```python
# 1. x1, x2, Z값 구하기
d = 100 # # of split 
x1 = np.linspace(0, 6, d) # 0~6
x2 = np.linspace(0, 10, d) #0~10
l1 = [3, 5]

Z = np.zeros((d,d)) # (100,100)
sigma = 1.0
for i,t1 in enumerate(x1):
    for j,t2 in enumerate(x2):
        val = GaussianKernel(np.array([t1,t2]), l1, sigma)
        Z[i,j] = val
```

```python
# 2. 그리기
from mpl_toolkits import mplot3d

pX, pY = np.meshgrid(x1, x2)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(pX, pY, Z, 100, cmap='jet')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Z')
plt.show()

#ax.view_init(60, 35)
fig
```

# 2. 선형 SVM 분류
# 1) 데이터 읽기

```python
import numpy as np

from scipy.io import loadmat
data = loadmat('data1.mat')
print data

X = data['X']
y = data['y']
m = len(y)
```

```python
print X.shape, y.shape
```

# 2) 그래프 그리기

```python
# y=1, y=0 데이터 인덱스 찾기
pos = []
neg = []

for (i, val) in enumerate(y):
    if val==1:
        pos.append(i)
    else:
        neg.append(i)
print pos
print neg
```

```python
import matplotlib.pyplot as plt
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label='y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')
plt.title("Data 1")
plt.show()
```

# 3) 학습

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
#평균0, 표준편차1로 만듦
from sklearn.svm import LinearSVC

svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_scv", LinearSVC(C=1, loss="hinge"))
])

svm_clf.fit(X, y)
```

```python
svm_clf.predict([[2.3,2.7],
                [3.7, 3.9]])
```

# 4) Decision boundary

```python
#Plot the decision boundary. For that, we will assign a color to each
#point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min()-0.2, X[:, 0].max()+0.2
y_min, y_max = X[:, 1].min()-0.2, X[:, 1].max()+0.2
h = .02 #step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])

#Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

#Plot also the training points
plt.plot(X[pos,0].reshape(-1), X[pos,1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg,0].reshape(-1), X[neg,1].reshape(-1), 'ro', label='y=0')
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```

# 3. 비선형 SVM 분류

```python
import numpy as np

#(x1),(x2),(0 or 1)
#Ng, Machine Learning, Coursera, ml-ex6 중
from scipy.io import loadmat
data = loadmat('data2.mat')
print data

X = data['X'] # x1, x2 의 값
y = data['y'] # 0 or 1
m = len(y)
```

```python
print X.shape, y.shape
```

# 2) 그래프 그리기

```python
# y=1, y=0 데이터 인덱스 찾기
pos = []
neg = []

for (i, val) in enumerate(y):
    if val==1:
        pos.append(i)
    else:
        neg.append(i)
print pos
print neg
```
